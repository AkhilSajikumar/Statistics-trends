```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#read the csv file into a Dataframe
datas=pd.read_csv('Song.csv')
```

```python
#Display all rows of DataFrame
datas
```

```python
#Checking null values
datas.isnull().sum()
```

```python
#Checking duplicate values
datas[datas.duplicated()]
```

```python
#Displaying basic staticstical functions
datas.describe()
```

```python
#Reducing the data to 80 rows
trend_song=datas.iloc[0:80,0:10]
```

```python
#Displaying all data's in 80 rows
trend_song
```

```python
#Creating a new figure with size 12, 12
plt.figure(figsize=(12, 12))
#plot a bar chart where the X axis represent Artist anf Y axis represent Rating
plt.bar(trend_song['Artist'],trend_song['Rating'])
plt.xlabel("Artist")
plt.ylabel("Rating")
plt.xticks(rotation=90, fontsize=8)
plt.title("Artist V/S Rating")
#Display the plot
plt.show()
```

```python
#Create a scatter plot with sales and streams
plt.scatter(trend_song["Sales"], trend_song["Streams"])
plt.xlabel("Sales")
plt.ylabel("Streams")
plt.title("Relation between Sales and Streams")
plt.show()
```

```python
trend_song
```

```python
#Checking the correlation
trend=trend_song[['Year', 'Sales', 'Streams', 'Downloads', 'Radio Plays', 'Rating'] ].corr()
trend
```

```python
#Calculating the median for song's Rating
medians=trend_song['Rating'].median()
medians
```

```python
#Calculating standard deviation for sales
Standard_deviation=trend_song['Sales'].std()
Standard_deviation
```

```python
#Calculating skew value of streams
skew=trend_song['Streams'].skew()
skew
```

```python
#Calculating kurtosis of downloads
Kurtosis=trend_song["Downloads"].kurtosis()
Kurtosis
```
